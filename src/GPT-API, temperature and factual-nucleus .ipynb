### Factual-nucleus sampling:

Гиперпараметры из статьи

```python
LAMBDA_DECAY = 0.9
P_RESET = 0.9
W_BOUND = 0.3
```

```python
end_of_sentence_idxes = ['.', '!', '?']
```

```python
output_texts = []
prompt = prompts_meanings[3]
    
cur_pos_in_sentence=1

cur_ans = ''
chat_completion = client.chat.completions.create(
    model="gpt-3.5-turbo",
    messages = [      
    {
        "role": "user",
        "content": prompt,
    },
    {'role':'assistant',
     'content': cur_ans}],
    max_tokens=1)

last_token = chat_completion.choices[0].message.content


for _ in range(20):
    nucleus_p = max(W_BOUND, P_RESET * pow(LAMBDA_DECAY, cur_pos_in_sentence - 1))

    chat_completion = client.chat.completions.create(
    model="gpt-3.5-turbo",
    messages=[
        
        {
            "role": "user",
            "content": prompt,
        },
        {'role':'assistant',
         'content': cur_ans}
    ],

    top_p = nucleus_p,
    max_tokens=1     
    )    

    last_token = chat_completion.choices[0].message.content

    if last_token in end_of_sentence_idxes:
        cur_pos_in_sentence = 1
    else:
        cur_pos_in_sentence += 1

    cur_ans += last_token

output_texts.append(cur_ans)
```

```python
from tqdm import tqdm_notebook

output_texts = []
for prompt in tqdm_notebook(prompts_meanings):
    
    cur_pos_in_sentence=1

    cur_ans = ''

    for _ in range(50):
        nucleus_p = max(W_BOUND, P_RESET * pow(LAMBDA_DECAY, cur_pos_in_sentence - 1))

        chat_completion = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=[

            {
                "role": "user",
                "content": prompt,
            },
            {'role':'assistant',
             'content': cur_ans}
        ],

        top_p = nucleus_p,
        max_tokens=3     
        )    

        last_token = chat_completion.choices[0].message.content

        dot_pos = last_token.find('.')
        if  dot_pos != -1:
            cur_pos_in_sentence = 1
        else:
            cur_pos_in_sentence += 1

        if dot_pos == -1:
            cur_ans += last_token + ' '
        else:
            cur_ans += last_token[:dot_pos+1] + ' '

    output_texts.append(cur_ans)
```
